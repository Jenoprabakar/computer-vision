```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('download.jpeg', cv2.IMREAD_GRAYSCALE)

if image is None:
    raise FileNotFoundError("Image not found. Check the file path.")


avg_3 = cv2.blur(image, (3, 3))
avg_5 = cv2.blur(image, (5, 5))
avg_7 = cv2.blur(image, (7, 7))


gauss_3 = cv2.GaussianBlur(image, (3, 3), sigmaX=0.5)#1/2 pi(sigma)
gauss_5 = cv2.GaussianBlur(image, (5, 5), sigmaX=1.0)
gauss_7 = cv2.GaussianBlur(image, (7, 7), sigmaX=2.0)


sobel_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)
sobel = cv2.magnitude(sobel_x, sobel_y)


prewitt_x = np.array([[-1, 0, 1],
                      [-1, 0, 1],
                      [-1, 0, 1]])
prewitt_y = np.array([[1, 1, 1],
                      [0, 0, 0],
                      [-1, -1, -1]])

prewittX = cv2.filter2D(image, -1, prewitt_x)
prewittY = cv2.filter2D(image, -1, prewitt_y)
prewitt = cv2.add(prewittX, prewittY)

laplacian = cv2.Laplacian(image, cv2.CV_64F)


median = cv2.medianBlur(image, 5)

min_filter = cv2.erode(image, np.ones((3, 3), np.uint8))
max_filter = cv2.dilate(image, np.ones((3, 3), np.uint8))


titles = [
    'Original Image',
    'Average Filter 3x3', 'Average Filter 5x5', 'Average Filter 7x7',
    'Gaussian σ=0.5', 'Gaussian σ=1.0', 'Gaussian σ=2.0',
    'Sobel Edge', 'Prewitt Edge', 'Laplacian',
    'Median Filter', 'Min Filter', 'Max Filter'
]

images = [
    image,
    avg_3, avg_5, avg_7,
    gauss_3, gauss_5, gauss_7,
    sobel, prewitt, laplacian,
    median, min_filter, max_filter
]

plt.figure(figsize=(15, 12))
for i in range(len(images)):
    plt.subplot(4, 4, i + 1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('download.jpeg', cv2.IMREAD_GRAYSCALE)

if image is None:
    raise FileNotFoundError("Image not found. Check the file path.")

avg_3 = cv2.blur(image, (3, 3))
avg_5 = cv2.blur(image, (5, 5))
avg_7 = cv2.blur(image, (7, 7))

gauss_3 = cv2.GaussianBlur(image, (3, 3), sigmaX=0.5)
gauss_5 = cv2.GaussianBlur(image, (5, 5), sigmaX=1.0)
gauss_7 = cv2.GaussianBlur(image, (7, 7), sigmaX=2.0)

sobel_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)
sobel = cv2.magnitude(sobel_x, sobel_y)

prewitt_x = np.array([[-1, 0, 1],
                      [-1, 0, 1],
                      [-1, 0, 1]])

prewitt_y = np.array([[1, 1, 1],
                      [0, 0, 0],
                      [-1, -1, -1]])

prewittX = cv2.filter2D(image, -1, prewitt_x)
prewittY = cv2.filter2D(image, -1, prewitt_y)
prewitt = cv2.add(prewittX, prewittY)

laplacian = cv2.Laplacian(image, cv2.CV_64F)

median = cv2.medianBlur(image, 5)
min_filter = cv2.erode(image, np.ones((3, 3), np.uint8))
max_filter = cv2.dilate(image, np.ones((3, 3), np.uint8))

titles = [
    'Original Image',
    'Average 3x3', 'Average 5x5', 'Average 7x7',
    'Gaussian σ=0.5', 'Gaussian σ=1.0', 'Gaussian σ=2.0',
    'Sobel', 'Prewitt', 'Laplacian',
    'Median', 'Min Filter', 'Max Filter'
]

images = [
    image,
    avg_3, avg_5, avg_7,
    gauss_3, gauss_5, gauss_7,
    sobel, prewitt, laplacian,
    median, min_filter, max_filter
]

plt.figure(figsize=(15, 12))
for i in range(len(images)):
    plt.subplot(4, 4, i + 1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.axis('off')

plt.tight_layout()
plt.show()

hist_titles = [
    'Histogram - Original',
    'Histogram - Average 5x5',
    'Histogram - Gaussian σ=1.0',
    'Histogram - Median',
    'Histogram - Sobel',
    'Histogram - Prewitt',
    'Histogram - Laplacian'
]

hist_images = [
    image,
    avg_5,
    gauss_5,
    median,
    sobel,
    prewitt,
    laplacian
]

plt.figure(figsize=(14, 10))
for i in range(len(hist_images)):
    plt.subplot(4, 2, i + 1)
    plt.hist(hist_images[i].ravel(), bins=256, range=(0, 256))
    plt.title(hist_titles[i])
    plt.xlabel('Pixel Intensity')
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
