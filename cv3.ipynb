```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread("uneven.jpg", cv2.IMREAD_GRAYSCALE)

_, manual = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)

_, otsu = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

adaptive_mean = cv2.adaptiveThreshold(
    img, 255, cv2.ADAPTIVE_THRESH_MEAN_C,
    cv2.THRESH_BINARY, 11, 2
)

gaussian_11 = cv2.adaptiveThreshold(
    img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
    cv2.THRESH_BINARY, 11, 2
)

gaussian_21 = cv2.adaptiveThreshold(
    img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
    cv2.THRESH_BINARY, 21, 2
)

gaussian_31 = cv2.adaptiveThreshold(
    img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
    cv2.THRESH_BINARY, 31, 2
)

titles = [
    "Original Grayscale",
    "Manual Threshold (127)",
    "Otsu's Threshold",
    "Adaptive Mean (11)",
    "Gaussian Adaptive (11)",
    "Gaussian Adaptive (21)",
    "Gaussian Adaptive (31)"
]

images = [
    img, manual, otsu,
    adaptive_mean, gaussian_11,
    gaussian_21, gaussian_31
]

plt.figure(figsize=(14, 10))
for i in range(len(images)):
    plt.subplot(3, 3, i+1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.axis('off')

plt.tight_layout()
plt.show()
```
